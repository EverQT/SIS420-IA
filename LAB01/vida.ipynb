```python
#import numpy as np

def featureNormalize(X):
  X = np.array(X, dtype=np.float64)
  X_norm = X.copy()
  mu = np.zeros(X.shape[1])

  mu = np.mean(X, axis = 0)
  sigma = np.std(X, axis = 0)
  X_norm = (X - mu)/ sigma

  return X_norm, mu, sigma
```

```python
X_norm, mu, sigma = featureNormalize(X)
print(X_norm)
```

```python
m, n = X_norm.shape
X = np.concatenate([np.ones((m, 1)), X_norm], axis=1)
```

```python
def calcularCosto(X, y, theta):
  m = y.size
  J = 0
  J = (1 /(2 * m)) * np.sum(np.square(np.dot(X, theta) - y))

  return J

#def calcular_costo(theta, X, y):
#  m = len(y)
#  h = X.dot(theta)
#  errores = h - y
#  J = (1 / (2 * m)) * np.dot(errores.T, errores)
#  return J.item()
#
```

```python
y = y.flatten() # Convierte un arreglo multidimensional en un vector unidimensional
```

```python
def gradientDescent(X, y, theta, alpha, num_iters):
  m = y.shape[0]
  theta = theta.copy()

  J_history = []

  for i in range(num_iters):
    theta = theta - (alpha / m) * (np.dot(X, theta) - y).dot(X)
    J_history.append(calcularCosto(X, y, theta))

  return theta, J_history
```

```python
# inicializa los parametros de ajuste
theta = np.zeros(2)

# configuraciones para el descenso por el gradiente
iterations = 500000
alpha = 0.001

theta, J_history = gradientDescent(X ,y, theta, alpha, iterations)
print('Theta encontrada por descenso gradiente: {:.4f}, {:.4f}'.format(*theta))
print('Valores theta esperados (aproximadamente): [-3.6303, 1.1664]')
print(J_history)
```

```python
# grafica la linea de ajuste
plotData(X[:, 1], y)
plt.plot(X[:, 1], np.dot(X, theta), '-')
plt.legend(['Datos de entrenamiento', 'RegresiÃ³n linear']);
```
